# DeepAR

```python
i=1
month=i
keys=['actual_p_max','unnecessary'] #,'accumated_neg','hourly_net_load_error'
postfix=''
track_necessary=4+6*(month-1)
compensation=5+6*(month-1)
track_necessary_compensation=6+6*(month-1)
merged_df_Jan=get_merged_df(
    file_folder=r'L:\Coding_project\Energy_grid_new_exp_local\Multiple_model_compare',
    log_fn=r"L:\Coding_project\Energy_grid_new\output\experiments\Multiple_model_compare\DeepAR_optuna.xlsx",
    id_exe_unne=17+5, id_sol_nece=161+5, id_mini=161+5
)
last=0
for i in merged_df_Jan.index:
    if merged_df_Jan.loc[i,'net_load_error']<0:
        last+=merged_df_Jan.loc[i,'net_load_error']
    merged_df_Jan.loc[i,'accumated_neg']=last
merged_df_Jan['accumated_neg']=-merged_df_Jan['accumated_neg']/30

# cal the hourly peak neg error
i=0
peak_neg_error=0
merged_df_Jan=merged_df_Jan.reset_index()
acc_step=4
while i < len(merged_df_Jan.index)-1:
    #k=0
    neg_sum=0
    for j in range(0,acc_step):
        try:
            neg_sum+=min(0,merged_df_Jan.loc[i+j,'net_load_error'])
        except:
            ...
    peak_neg_error=min(neg_sum,peak_neg_error)
    for j in range(0,acc_step):
        try:
            merged_df_Jan.loc[i+j,'hourly_net_load_error']=peak_neg_error
        except:
            ...
    i=i+acc_step
last=0
i=0
while i < len(merged_df_Jan.index)-1:
    if merged_df_Jan.loc[i,'unnecessary']<last or merged_df_Jan.loc[i,'unnecessary']==None:
        merged_df_Jan.loc[i,'increasing']=1
    else:
        merged_df_Jan.loc[i,'increasing']=0
    last=merged_df_Jan.loc[i,'unnecessary']
    i+=1
merged_df_Jan.index=(pd.to_datetime(merged_df_Jan['index']))
merged_df_Jan['net_load_error']=merged_df_Jan['net_load_error']

line_clr_dic={
    'actual_p_max':'teal',
    'necessary':'green',
    'unnecessary':'coral',
    'unnecessary_minimized':'purple',
    "latest_max_neg_net_load_error":'red',
    'accumated_neg':'indianred',  
    'hourly_net_load_error':'crimson'   
    }
label_dic={
    'actual_p_max':'Peak demand till t',
    'necessary':'green',
    'unnecessary':'discr. w/GT',
    'unnecessary_minimized':'purple',
    "latest_max_neg_net_load_error":'Peak error- till t',
    'accumated_neg':'Normalized acc. error(-) ',
    'hourly_net_load_error':'Peak hourly error- till t',
    'gray':'Forecast error (right axis)',
    'red':'Forecast error at suboptimal (right axis)'
    }

plot_track_p_max(df_merged=merged_df_Jan,figsize=(12,6),
                line_keys=keys,plot_error_bar=True,month='Jan',
                ylimit_main=[-200,300],ylimit_sub=[-100,200],
                linewidth=1.5, legend_loc="lower left", all_positive=False,
                ylabel='Power (kW)', xlabel_l='Day of month (Oct-2019)',xlabel_r='Hour of day (4-Oct to 6-Oct)',
                inside_start_day=3,inside_days=3,shadow=True,
                line_clr_dic=line_clr_dic,label_dic=label_dic,
                save_fn="test_"+str(i)+postfix+"100dpi.png")
```
